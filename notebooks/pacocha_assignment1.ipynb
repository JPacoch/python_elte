## Task 9

Write the `revenue_per_employee(company)` function, which receives a dictionary (or object) created in *Task 7*, then calculates and returns the revenue / employee for that company.
Call the function on each company in a loop and display the results!

````python
def revenue_per_employee(company):
    for dic in lista:
        try:
            if dic["name"] == company:
                re = dic["revenue"] / dic["employees"]
                dic["revenue/employees"] = re
                return dic
                break
        except:
            print('nan')


print(revenue_per_employee("General Mills"))

lista3 = []
lista2 = [obiekt['name'] for obiekt in lista]
for k in range(0, len(lista2)):
    lista3.append(revenue_per_employee(lista2[k]))
    k + 1
print(lista3)
````
